```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#força a saída dos gráficos (senão precisa usar plt.show() após cada plot)
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x * x
y
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,x**2, 'g--', linewidth=2)
ax.plot(x,x**3, color='#FFFF0F', linewidth=2)
ax.plot(x,x**4, color='pink', linewidth=5, alpha=0.5, linestyle=':')

ax.set_xlim([0,2])
ax.set_ylim([0,10])
```
