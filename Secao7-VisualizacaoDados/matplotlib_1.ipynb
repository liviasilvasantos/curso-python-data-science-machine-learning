```python
import matplotlib.pyplot as plt
```

```python
#força a saída dos gráficos (senão precisa usar plt.show() após cada plot)
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color='r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título do nosso gráfico')
```

```python
plt.plot(x,y, '--r')
```

```python
#subplots
#plt.subplot(nrows=1, ncols=2)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig = plt.figure()
#porcentagem de distância - esquerda, abaixo, largura, altura
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')

axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
axes2.plot(y, x)
```

```python
#cria duas instâncias e já cria um eixo
fig, ax = plt.subplots()
ax.plot(x,x**3, 'b--')
```

```python
#cria múltiplos eixos na mesma imagem
fig, ax = plt.subplots()
ax.plot(x,x**3, 'b--')
ax.plot(x,x**4, 'g--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('título')
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3)
```

```python
ax[0].plot(x,y)
ax[0].set_title('eixo o')
```

```python
fig = plt.figure(figsize=(8,4), dpi=100) 
```

```python
fig, axes = plt.subplots(figsize=(8,5))
axes.plot(x,y, 'r-')
axes.set_title('Título')
fig.savefig("imagem.png")
```

```python
fig, axes = plt.subplots(figsize=(8,5))
axes.plot(x,y, 'r-', label='x ˆ 2')
axes.plot(y,x, 'g--', label='x ** 1/2')
axes.set_title('Título')
axes.legend(loc=0)
```
